```python
import pandas as pd

#Importing Economic data
SocialData = pd.read_csv("C:/Users/denis/OneDrive/VSCode/Git/culminating-project-group-2/WEEK 3/Indicators/Social_Indicators.csv")

SocialData.head()
```

```python
# Import the tabulate module
from tabulate import tabulate

#identifying data types and missing values in one table
#Get the data types
dtypes = SocialData.dtypes

# Get the number of missing values
missing_values = SocialData.isna().sum()

#Creating a new DataFrame for combined information
info = pd.DataFrame({
    'Column': dtypes.index,
    'DataType': dtypes.values,
    'MissingValues': missing_values.values
})

#Converting DataFrame to table format with right alignment
table = tabulate(info, headers='keys', tablefmt='pretty', showindex=False, colalign=("right", "right", "right"))

#Display the table
print(table)
```
