```python
import pandas as pd

# Load the datasets
env_df = pd.read_csv('/content/Envirnmental_Indicators.csv')
eco_df = pd.read_csv('/content/economic_indicators_wdi.csv')

# View first few rows
print("Environmental Indicators:")
display(env_df.head())

print("\nEconomic Indicators:")
display(eco_df.head())

# Quick info
print("\nEnvironmental Indicators Info:")
print(env_df.info())

print("\nEconomic Indicators Info:")
print(eco_df.info())

# Quick summary
print("\nEnvironmental Indicators Summary:")
display(env_df.describe())

print("\nEconomic Indicators Summary:")
display(eco_df.describe())
```

```python
import numpy as np

# Replace '..' and '...' with NaN
env_df.replace(['..', '...'], np.nan, inplace=True)
eco_df.replace(['..', '...'], np.nan, inplace=True)

# Convert columns to numeric where needed (for important columns)
important_env_cols = [
    'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]',
    'Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita) [EN.GHG.CO2.PC.CE.AR5]',
    'Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]',
    'Access to electricity, rural (% of rural population) [EG.ELC.ACCS.RU.ZS]',
    'Access to electricity, urban (% of urban population) [EG.ELC.ACCS.UR.ZS]',
    'Forest area (sq. km) [AG.LND.FRST.K2]'
]

for col in important_env_cols:
    env_df[col] = pd.to_numeric(env_df[col], errors='coerce')

# Same for economic important columns
important_eco_cols = [
    'GDP (current US$) [NY.GDP.MKTP.CD]',
    'GDP per capita (current US$) [NY.GDP.PCAP.CD]',
    'GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'
]

for col in important_eco_cols:
    eco_df[col] = pd.to_numeric(eco_df[col], errors='coerce')


```

```python
print("Environmental dataset missing values:")
print(env_df[important_env_cols].isnull().sum())

print("\nEconomic dataset missing values:")
print(eco_df[important_eco_cols].isnull().sum())
```

```python
env_df_clean = env_df.dropna(subset=[
    'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]',
    'Carbon dioxide (CO2) emissions excluding LULUCF per capita (t CO2e/capita) [EN.GHG.CO2.PC.CE.AR5]'
])

eco_df_clean = eco_df.dropna(subset=[
    'GDP (current US$) [NY.GDP.MKTP.CD]',
    'GDP per capita (current US$) [NY.GDP.PCAP.CD]',
    'GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'
])
```

```python
# Replace '..' and '...' with NaN if you haven't already
import numpy as np
env_df.replace(['..', '...'], np.nan, inplace=True)
eco_df.replace(['..', '...'], np.nan, inplace=True)

# Convert important columns to numeric
important_env_cols = [
    'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]',
    'Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]'
]

important_eco_cols = [
    'GDP (current US$) [NY.GDP.MKTP.CD]',
    'GDP per capita (current US$) [NY.GDP.PCAP.CD]',
    'GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'
]

for col in important_env_cols:
    env_df[col] = pd.to_numeric(env_df[col], errors='coerce')

for col in important_eco_cols:
    eco_df[col] = pd.to_numeric(eco_df[col], errors='coerce')

# Merge datasets first
merged_df = pd.merge(env_df, eco_df, on=["Country Name", "Time"], how="inner")

# Now drop rows ONLY IF all important variables are missing
merged_df_clean = merged_df.dropna(subset=[
    'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]',
    'Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]',
    'GDP per capita (current US$) [NY.GDP.PCAP.CD]',
    'GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'
], how='all')

print(merged_df_clean.shape)
merged_df_clean.head()
```

```python
# Drop rows with missing values in key columns
merged_df_clean = merged_df_clean.dropna(subset=[
    'Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]',
    'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]',
    'GDP per capita (current US$) [NY.GDP.PCAP.CD]',
    'GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'
])

# Check the new shape
print("Shape after dropping NaNs:", merged_df_clean.shape)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(
    merged_df_clean['Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]'],
    merged_df_clean['Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]']
)
plt.xlabel('Renewable Energy Consumption (%)')
plt.ylabel('CO₂ Emissions (Mt)')
plt.title('Renewable Energy Consumption vs CO₂ Emissions')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.scatter(
    merged_df_clean['GDP per capita (current US$) [NY.GDP.PCAP.CD]'],
    merged_df_clean['Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]']
)
plt.xlabel('GDP Per Capita (US$)')
plt.ylabel('CO₂ Emissions (Mt)')
plt.title('GDP Per Capita vs CO₂ Emissions')
plt.grid(True)
plt.show()
```

```python
import seaborn as sns

# Focused on important columns
corr_cols = [
    'Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]',
    'Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]',
    'GDP (current US$) [NY.GDP.MKTP.CD]',
    'GDP per capita (current US$) [NY.GDP.PCAP.CD]',
    'GDP growth (annual %) [NY.GDP.MKTP.KD.ZG]'
]

# Calculate correlation matrix
corr_matrix = merged_df_clean[corr_cols].corr()

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title('Correlation Matrix of Key Variables')
plt.show()
```

```python
# Group by Time and calculate mean only for numeric columns
time_trends = merged_df_clean.groupby('Time').mean(numeric_only=True)

# Now plot Renewable Energy Consumption trend
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.plot(time_trends.index, time_trends['Renewable energy consumption (% of total final energy consumption) [EG.FEC.RNEW.ZS]'], marker='o')
plt.xlabel('Year')
plt.ylabel('Renewable Energy Consumption (%)')
plt.title('Average Renewable Energy Consumption Over Years')
plt.grid(True)
plt.show()
```

```python
# Plot CO2 Emissions
plt.figure(figsize=(8,6))
plt.plot(time_trends.index, time_trends['Carbon dioxide (CO2) emissions (total) excluding LULUCF (Mt CO2e) [EN.GHG.CO2.MT.CE.AR5]'], marker='o')
plt.xlabel('Year')
plt.ylabel('CO₂ Emissions (Mt)')
plt.title('Average CO₂ Emissions Over Years')
plt.grid(True)
plt.show()
```
